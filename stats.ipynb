```python
import json
import matplotlib.pyplot as plt
from collections import defaultdict
from tqdm import tqdm
import seaborn as sns
sns.set_theme()
```

```python
artists = json.loads(open("data/spotify_artists.json").read())
```

```python
genres = defaultdict(lambda: 0)
for artist in artists:
    for genre in artist["genres"]:
        genres[genre] += 1

genres_list = sorted([(x,y) for x,y in genres.items()], 
                     key=lambda x: x[1], 
                     reverse=True
                     )
n=20
topn = genres_list[:n]
genre = [x[0] for x in topn]
count = [x[1] for x in topn]


plt.figure(figsize=(20, 10))
plt.bar(genre, count)
plt.xticks(rotation=90)
plt.show()
```
